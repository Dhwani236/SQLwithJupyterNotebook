```python
import sqlite3 
import pandas as pd 
from db_setup import setup_database
setup_database()

conn=sqlite3.connect('bookcycle.db')
```

```python
Learners_AI_Response="""
To find the most preferred payment method grouped by each birth year (for customers born before 2000), you need to rank the payment methods by popularity within each birth year group. Since SQLite supports window functions, you can achieve this efficiently using a Common Table Expression (CTE) and ROW_NUMBER():

SELECT c.birth_year, t.payment_method, COUNT(t.payment_method) 
FROM customers c 
NATURAL JOIN transactions t 
GROUP BY t.payment_method;

## How It Works

* COUNT(DISTINCT t.customer_id): This ensures we count the number of unique people using that payment method per birth year, rather than just counting raw transaction slips (in case one person bought 50 books with the same card).
* PARTITION BY c.birth_year: The window function isolates each birth year (e.g., 1994) as its own separate group.
* ORDER BY ... DESC: This sorts the payment methods within that birth year group from most popular to least popular.
* WHERE rank_num = 1: The final filter strips away the less popular choices, leaving you with exactly one top-preferred payment method for every single birth year.

## Handling Ties
If two payment methods are equally popular for a specific year and you want to see both, change ROW_NUMBER() to DENSE_RANK() in the query above.
If you would like to test this on specific years or see how to include total revenue alongside the count, let me know!

"""
display("Your AI-Generated Query:")
display(Learners_AI_Response)
```
